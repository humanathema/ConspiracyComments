# Pipeline Step 01: Data Ingestion

This notebook imports our clean logic from `src/ingestion.py` and runs the core data parsing.
**Caching Guarantee**: It will check if the outputs already exist in `data/processed/` before running any heavy extraction.

```python
import os
import pandas as pd
import sys

# Ensure we can import from src/
sys.path.append(os.path.abspath(os.path.join("..", "..")))
from src.ingestion import extract_media_titles

BASE_DIR = "/Users/nash/Projects/ConspiracyComments/"
RAW_POSTS = os.path.join(BASE_DIR, "data/raw/r_conspiracy_posts.jsonl")
PROCESSED_MEDIA = os.path.join(BASE_DIR, "data/processed/media_titles.csv")

print("Ready.")
```

## 1. Extract Media Titles (Cached)

```python
if os.path.exists(PROCESSED_MEDIA):
    print(f"✅ Cache found! Loading directly from {PROCESSED_MEDIA}")
    media_df = pd.read_csv(PROCESSED_MEDIA)
    print(f"Loaded {len(media_df):,} records.")
else:
    print(f"⚠️ Cache not found. Running massive JSONL extraction...")
    media_df = extract_media_titles(RAW_POSTS, target_limit=10000)
    media_df.to_csv(PROCESSED_MEDIA, index=False)
    print(f"✅ Extraction complete and saved to cache!")
    
media_df.head()
```
